```python
import numpy as np

def calculate(list):

  if len(list) < 9 or len(list) > 9:
    #return "List must contain nine numbers."
    raise ValueError("List must contain nine numbers.")

  else:
    list = np.array(list).reshape(3, 3)
    print(list, "\n")

    
    new_dict = dict()
    
    a = np.mean(list, axis=0).tolist()
    b = np.mean(list, axis=1).tolist()
    new_dict['mean'] = [a, b, np.mean(list)]
    
    a = np.var(list, axis=0).tolist()
    b = np.var(list, axis=1).tolist()
    new_dict['variance'] = [a, b, np.var(list)]
    
    a = np.std(list, axis=0).tolist()
    b = np.std(list, axis=1).tolist()
    c = np.std(list)
    new_dict['standard deviation'] = [a, b, c]
    
    a = np.max(list, axis=0).tolist()
    b = np.max(list, axis=1).tolist()
    c = np.max(list)
    new_dict['max'] = [a, b, c]
    
    a = np.min(list, axis=0).tolist()
    b = np.min(list, axis=1).tolist()
    c = np.min(list)
    new_dict['min'] = [a, b, c]
    
    a = np.sum(list, axis=0).tolist()
    b = np.sum(list, axis=1).tolist()
    c = np.sum(list)
    new_dict['sum'] = [a, b, c]
    
    # YES!
    # https://stackoverflow.com/questions/60051461/how-to-get-numpy-array-values
    
    #my_dict = {"mean":[a.tolist()],"variance":[],"std":[], 'max':[], 'min':[], 'sum':[]};
    #print("\n :(((((", my_dict, "\n\n")
    
    print(np.mean(list, axis=0), "\n")
    

    return new_dict


print(calculate([0,1,2,3,4,5,6,7,8]))
```
